**Question 1:** How many customers are in the file?
(as help, we have added some comments and starter code to help you structure the solution)

```python
from os.path import exists
import pandas as pd
import csv

# check if file can be found and open the file
datafile = "./sample_data/customers.csv"
if not exists(datafile):
  raise SystemExit("You should run the first code cell and download the dataset files!")

# customers file is in the sample_data folder and can be opened.
# Using csv reader is one way of dealing with the file
# but it's up to you how you want to proceed


customers_df = pd.read_csv(datafile)
print(customers_df.head())

customers_df = customers_df.dropna(subset=['CustomerID'])
print(customers_df.dtypes)
customers_df['CustomerID'] = customers_df['CustomerID'].astype(int)
customers_df = customers_df.drop_duplicates(subset=['CustomerID'])

customers_df['State'] = customers_df['State'].str.strip()
customers_df['State'] = customers_df['State'].str.strip().str.upper()


customers_count = customers_df['CustomerID'].nunique()
print(f"Total number of customers: {customers_count}")

print("All done!")
```

**Question 2:** In how many different states do the customers live in?

```python
# open file and init vars
# evaluate the contents of the file

states_count = customers_df['State'].nunique()
print(f"Total number of states: {states_count}")
```

**Question 3** What is the state with most customers?

```python
# open file and init vars
# evaluate the contents of the file

state_customers = customers_df.groupby('State').size().sort_values(ascending=False)
print(state_customers)
most_customers_state = state_customers.index[0]
count_most_customers = state_customers.iloc[0]
print(f"The state with most customers is {most_customers_state} with {count_most_customers} customers" )
```

**Question 4** What is the state with the least customers?

```python
# open file and init vars
# evaluate the contents of the file

min_count = state_customers.min()
least_customers_states = state_customers[state_customers == min_count]
print(f"The states with the least customers ({min_count}) are:")
print(least_customers_states.index.tolist())
```

**Question 5:** What is the most common last name?

```python
# open file and init vars
# evaluate the contents of the file

last_name_counts = customers_df['LastName'].value_counts().idxmax()
print(f"The most common last name is {last_name_counts}")
```

```python
# open and evaluate orders.csv

orders_datafile = "./sample_data/orders.csv"
if not exists(orders_datafile):
  raise SystemExit("You should run the first code cell and download the dataset files!")

orders_df = pd.read_csv(orders_datafile)
print(orders_df.head())
print(orders_df.dtypes)

orders_df = orders_df.dropna(subset=['CustomerID'])
orders_df['CustomerID'] = orders_df['CustomerID'].astype(int)
orders_df['Date'] = pd.to_datetime(orders_df['Date'])

```


**Question #1:** How many unique orders are in the orders.csv file?


```python

unique_orders_count = orders_df['OrderID'].nunique()
print(f"There are {unique_orders_count} orders")
```

**Question #2:** What is the average number of items per order (rounded to two decimal places)?

> Agregar bloque entrecomillado



```python
items_per_order = orders_df.groupby('OrderID').size()
average_items_per_order = round(items_per_order.mean(),2)
print(f"The average number of items per order is {average_items_per_order}")
```

**Question #3:** What is the highest number of items per order?


```python

highest_items_per_order = items_per_order.max()
print(f"The highest number of items per order is {highest_items_per_order}")
```

**Question #4:** What is the number of orders placed in October 2021?\

```python
unique_orders = orders_df.drop_duplicates(subset=['OrderID'])

orders_oct_2021 = unique_orders[(unique_orders['Date'].dt.year == 2021) & (unique_orders['Date'].dt.month == 10)]
print(f"The number of orders placed in October 2021 is {len(orders_oct_2021)}")
```

**Question #5:** Which customer spent the most amount of money in 2021?

```python
spend_2021 = unique_orders[unique_orders['Date'].dt.year == 2021].groupby('CustomerID')['OrderTotal'].sum()
top_customer_id = spend_2021.idxmax()
customer_info = customers_df[customers_df['CustomerID'] == top_customer_id].iloc[0]

print(f"The customer who spent the most amount of money in 2021 is {customer_info['FirstName']} {customer_info['LastName']}")
```

**Question #6:** Historically, what is the best month for sales?

```python
import calendar

unique_orders_dated = unique_orders.dropna(subset=['Date']).copy()
unique_orders_dated['Month'] = unique_orders_dated['Date'].dt.month

month_sales = unique_orders_dated.groupby('Month')['OrderTotal'].sum()
best_month_name = calendar.month_name[month_sales.idxmax()]
print(f"The best month for sales is {best_month_name}")
```
